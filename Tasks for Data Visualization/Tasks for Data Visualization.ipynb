```python
import pandas as pd
```

```python
df = pd.read_csv('covid_19_country_wise_latest.csv')
```

```python
#1
df.head()
```

```python
#2
df.set_index('Unnamed: 0', inplace=True)
df.head()
```

```python
# 3
df.index.name = "index"
```

```python
# 4
def clean_column_names(dataframe):
    dataframe.columns = dataframe.columns.str.replace(" ", "_")
    return dataframe

df = clean_column_names(df)

df
```

```python
# 5
column_info = df.dtypes
missing_values = df.isnull().sum()
missing_values
```

```python
# 6
df.fillna(df.mean(), inplace=True)
```

```python
# 7
unique_countries = df['Country/Region'].nunique()
unique_countries
```

```python
# 9
mean_cases = df['Confirmed'].mean()
median_cases = df['Confirmed'].median()
std_cases = df['Confirmed'].std()
mean_cases
```

```python
# 10
df['Deaths'] = pd.to_numeric(df['Deaths'], errors='coerce')
df['Deaths'].fillna(df['Deaths'].mean(), inplace=True)

df
```

```python
#11
df['Deaths'] = df['Deaths'].astype(int)
df
```

```python
# 12
total_deaths = df['Deaths'].sum()
total_recovered = df['Recovered'].sum()

total_deaths,total_recovered
```

```python
# 13
countries_over_million = df[df['Confirmed'] > 1_000_000]['Country/Region'].count()
countries_over_million
```

```python
# 14
high_recovery_countries = df[df['Recovered'] / df['Confirmed'] > 0.95]['Country/Region'].tolist()
high_recovery_countries
```

```python
# 15
df.drop(columns=['WHO_Region', 'Confirmed'], inplace=True)
df
```

```python
# 16
country_max_deaths = df.loc[df['Deaths'].idxmax(), 'Country/Region']
country_max_deaths
```

```python
# 17
df_sorted_by_deaths = df.sort_values(by='Deaths', ascending=False)
df_sorted_by_deaths
```

```python
# 18
df['Total_cases'] = df['Deaths'] + df['Recovered'] + df['Active']
df
```

```python
# 19 
df['Death_Rate'] = (df['Deaths'] / df['Total_cases']) * 100
df['Death_Rate']
```

```python
# 20
low_death_rate_countries = df[(df['1_week_change'] > 0) & (df['Death_Rate'] < 1)]['Country/Region'].tolist()
low_death_rate_countries
```

```python
# 21 
head_death_rate = df[['Country/Region', 'Death_Rate']].head()
head_death_rate
```

```python
# 22
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['Total_cases'], df['Deaths'], alpha=0.5)
plt.xlabel("Total Cases")
plt.ylabel("Total Deaths")
plt.title("Scatter Plot of Total Cases vs Total Deaths")
plt.grid(True)
plt.show()
```
